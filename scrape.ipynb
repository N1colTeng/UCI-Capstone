```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.edge.options import Options
from selenium.webdriver.edge.service import Service
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('Security Updates 2024-10-18-024154pm.csv')
df.head()
```

```python
df = df.drop('Platform', axis=1)
df.head()
```

```python
df = df.drop_duplicates(subset=['Details', 'CWE'])
df = df.reset_index(drop=True)
df.head()
```

```python
df = df.rename(columns={
    'Details': 'CVE',
    'Details (Link)': 'Link'
})
df.head()
```

```python
df['CVE_Link'] = 'https://www.cve.org/CVERecord?id=' + df['CVE']
df.head()
```

```python
df.shape
```

```python
def scrape(link):
    edge_driver_path = "C:/users/nicol/Desktop/edgedriver_win32/msedgedriver.exe"

    options = Options()
    options.add_argument("headless")
    options.add_argument("disable-gpu")

    service = Service(edge_driver_path)
    driver = webdriver.Edge(service=service, options=options)
    driver.get(link)

    time.sleep(2)

    page_source = driver.page_source

    soup = BeautifulSoup(page_source, 'html.parser')
    section = soup.find(id="cve-desciption")
    description_text = section.find('p').get_text(strip=True)
    
    driver.quit()
    
    return description_text
```

```python
df['Description'] = df['CVE_Link'].apply(scrape)
```
